## Chapter 7 - Filter out missing data

```python
# Filter out missing data on columns we consider important for analysis

df = df.dropna(subset=['Internal ID', 'Q2: GENDER', 'Q3: AGE', 'Q4: COUNTRY'], inplace = False)
```

```python
df.info()
```

## Chapter 7 - Remove Duplicates

```python
# Drop duplicates
df = df.drop_duplicates()
```

```python
df.info()
```

```python
# The info above shows now dups were removed. 
```

## Chapter 7 - Transform data (make column upper case)

```python
df['Q4: COUNTRY'] = df['Q4: COUNTRY'].str.upper()
```

```python
df.head()
```

## Chapter 8 - Combine and merge data sets

```python
# Load a scored data set I created
df2 = pd.read_csv('candy_score.csv')  
```

```python
df2.info()
```

```python
# merge the score data set with the original data set
df3 = pd.merge(df, df2, on='Internal ID')
```

```python
df3.info()
```

```python
# Note from the above info() that the new data set has 121 columns, 
# that is one more than the original because we added the score column
```

## Chapter 10 - Grouping

```python
# Select score, group by gender
score_by_gender = df3['Score'].groupby(df3['Q2: GENDER'])
score_by_gender.head()
```

```python
# Calculate mean
score_by_gender.mean()
```

```python
# From the above, we can see that Males scored a little higher than females in the candy survey.
```

## Chapter 10 - Fill missing values with group-specific values

```python
# We are missing some values for Q12: MEDIA [Science] using the groupped mean

fill_mean = lambda g: g.fillna(g.mean())
df3.groupby('Q12: MEDIA [Science]').apply(fill_mean)
```
